```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
input_image = cv.imread('Input/emma.jpg', cv.IMREAD_GRAYSCALE)  

t1 = np.linspace(0,50,51)
t2 = np.linspace(51,100,0)
t3 = np.linspace(100,255,100)
t4 = np.linspace(255,150,0)
t5 = np.linspace(151,255,105)
tf = np.concatenate((t1,t2,t3,t4,t5), axis=0).astype('uint8')

# Create the output image
output_image = cv.LUT(input_image, tf)

# Plot the intensity transformation graph
plt.plot(range(256), tf, label='Intensity Transformation')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.title('Intensity Transformation Graph')
plt.legend()

# Save the plot as an image
plt.savefig('Output/intensity_transformation_graph_Q1.png')


# Save or display the output image
cv.imwrite('Output/emma_output_Q1.png', output_image)  # Save the output image

```

```python
fig, ax = plt.subplots(1, 2, figsize=(6, 8))
ax[0].imshow(input_image, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Input Image')
ax[1].imshow(output_image, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Output Image')
for a in ax:
    a.axis('off')
plt.show()
```
